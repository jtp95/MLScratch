```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Synthetic Data

```python
def generate_synthetic_data(n_samples=200, random_seed=42):
    np.random.seed(random_seed)

    # Generate class 0 (negative class)
    x0 = np.random.randn(n_samples // 2, 2) + np.array([-2, -2])

    # Generate class 1 (positive class)
    x1 = np.random.randn(n_samples // 2, 2) + np.array([2, 2])

    # Combine both classes
    X = np.vstack((x0, x1))
    y = np.hstack((np.zeros(n_samples // 2), np.ones(n_samples // 2)))  # Labels: 0s and 1s

    # Shuffle the dataset
    indices = np.arange(n_samples)
    np.random.shuffle(indices)
    X, y = X[indices], y[indices]

    # Convert to Dataframe
    df = pd.DataFrame(X, columns=["Feature1", "Feature2"])
    df["Label"] = y.astype(int)
    
    # Save to CSV
    csv_filename = "synthetic_data.csv"
    df.to_csv(csv_filename, index=False)
    print(f"Dataset saved to {csv_filename}")

    return X, y
```

```python
# Generate data
X, y = generate_synthetic_data(n_samples=300)

# Plot the data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', alpha=0.8)
plt.xlabel("Feature 1 (x1)")
plt.ylabel("Feature 2 (x2)")
plt.title("Synthetic Data for Logistic Regression")
plt.colorbar(label="Class Label (0 or 1)")
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer

# Load dataset
data = load_breast_cancer()

# Convert to DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)
df['Label'] = data.target  # 0 = Benign, 1 = Malignant

# Save dataset to CSV (Optional)
df.to_csv("breast_cancer_data.csv", index=False)

# Print dataset info
print("Dataset Shape:", df.shape)
print("Class Distribution:\n", df["Label"].value_counts())
```
